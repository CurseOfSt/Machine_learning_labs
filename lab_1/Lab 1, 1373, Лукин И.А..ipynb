```python
# Данный датасет был взят с сайта "https://www.kaggle.com/datasets/sujaykapadnis/cats-vs-dogs". Датасет представляет собой данные по содержанию кошек и собак в домах США.
# Данные взяты с "https://www.avma.org/resources-tools/reports-statistics/us-pet-ownership-statistics". 
# Атрибуты: 
#     числовые: номер, количество домов, процент домов с домашними животными, количество домов с питомцами, процент домов собачников, количество домов собачников, среднее количество собак на дом, популяция собак, процент кошатников, количество домов кошатников, среднее количество кошек на дом, популяция кошек;
#     строковые:  название штата.
# Гипотетические задачи анализа: в каких штатах домашнему питомцу наиболее комфортно жить.
```

```python
import numpy as np
import pandas as pd
df = pd.read_csv('cats_vs_dogs.csv')
df
```

```python
# вычислим среднее значение 
print(df['n_households'].mean())
print(df['percent_pet_households'].mean())
print(df['n_pet_households'].mean())
print(df['percent_dog_owners'].mean())
print(df['n_dog_households'].mean())
print(df['avg_dogs_per_household'].mean())
print(df['dog_population'].mean())
print(df['percent_cat_owners'].mean())
print(df['n_cat_households'].mean())
print(df['avg_cats_per_household'].mean())
print(df['cat_population'].mean())
```

```python
# вычислим СКО
sum=0
for i in df['n_households']:
    if (i!=np.nan):
        sum+=(abs(df['n_households'].mean()-i))**2
print(sum/df['n_households'].value_counts())
sum=0
for i in df['percent_pet_households']:
    if (i!=np.nan):
        sum+=(abs(df['percent_pet_households'].mean()-i))**2
print(sum/df['percent_pet_households'].value_counts())
sum=0
for i in df['n_pet_households']:
    if (i!=np.nan):
        sum+=(abs(df['n_pet_households'].mean()-i))**2
print(sum/df['n_pet_households'].value_counts())
sum=0
for i in df['percent_dog_owners']:
    if (i!=np.nan):
        sum+=(abs(df['percent_dog_owners'].mean()-i))**2
print(sum/df['percent_dog_owners'].value_counts())
sum=0
for i in df['n_dog_households']:
    if (i!=np.nan):
        sum+=(abs(df['n_dog_households'].mean()-i))**2
print(sum/df['n_dog_households'].value_counts())
sum=0
for i in df['avg_dogs_per_household']:
    if (i!=np.nan):
        sum+=(abs(df['avg_dogs_per_household'].mean()-i))**2
print(sum/df['avg_dogs_per_household'].value_counts())

sum=0
for i in df['dog_population']:
    if (i!=np.nan):
        sum+=(abs(df['dog_population'].mean()-i))**2
print(sum/df['dog_population'].value_counts())
sum=0
for i in df['percent_cat_owners']:
    if (i!=np.nan):
        sum+=(abs(df['percent_cat_owners'].mean()-i))**2
print(sum/df['percent_cat_owners'].value_counts())
sum=0
for i in df['n_cat_households']:
    if (i!=np.nan):
        sum+=(abs(df['n_cat_households'].mean()-i))**2
print(sum/df['n_cat_households'].value_counts())
sum=0
for i in df['avg_cats_per_household']:
    if (i!=np.nan):
        sum+=(abs(df['avg_cats_per_household'].mean()-i))**2
print(sum/df['avg_cats_per_household'].value_counts())
sum=0
for i in df['cat_population']:
    if (i!=np.nan):
        sum+=(abs(df['cat_population'].mean()-i))**2
print(sum/df['cat_population'].value_counts())
```

```python
# Пропущенные значения отсутсвуют, поэтому обработки не требуют, но в случае их наличия можно:
# 1)удалить строки или столбцы с пропущенными значениями. Это подходит, если пропущенные значения являются незначительными и удаление не влияет на анализ данных;
# 2)заполнить пропущенные значения средним или медианным значением атрибута;
# 3)для числовых данных можно использовать метод интерполяции для предсказания пропущенных значений на основе соседних значений.
```

```python
n_df=df.drop(columns=['state','Unnamed: 0'])
n_df_corr_matr = n_df.corr()
n_df_corr_matr
```

```python
сorrelation_koef = 0.25
uncorrelated_pairs = [[i, j] for i in range(n_df_corr_matr.shape[0])
                                 for j in range(n_df_corr_matr.shape[1])
                                     if (abs(n_df_corr_matr.iloc[i, j]) < сorrelation_koef)]
for pairs in uncorrelated_pairs:
    i,j = pairs
    n_df_corr_matr.iloc[i][j] = 0
n_df_corr_matr
```
